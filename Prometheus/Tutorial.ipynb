prometheus --config.file=Prometheus/prometheus.yml   

config.file 写配置文件(目标监控端口等)

当前累计请求数：
"namespace_name"_requests_total
下面以 myapp_api_requests_total 为例子

最近一分钟 QPS：
sum(rate(myapp_api_requests_total[1m]))

最近五分钟请求增加量：
increase(myapp_api_requests_total[5m])

错误率：
sum(rate(myapp_api_requests_total{status="error"}[5m]))
/
sum(rate(myapp_api_requests_total[5m]))

平均延迟：
sum(rate(request_latency_seconds_sum[5m]))
/
sum(rate(request_latency_seconds_count[5m]))

P95:
histogram_quantile(
  0.95,
  sum by (le) (
    rate(request_latency_seconds_bucket[5m])
  )
)

histogram_quantile(
  0.95,
  sum by (le, endpoint) (
    rate(request_latency_seconds_bucket[5m])
  )
)
